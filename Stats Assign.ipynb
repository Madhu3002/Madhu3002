```python
import os
import numpy as np
import pandas as pd
```

```python
af=pd.read_csv('airfare.csv')
```

```python
af.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
plt.figure(figsize=(10,8))
sb.lineplot(data=af)
```

### Arrival avg population is lesser than the Departure avg population.
### Fare is nominal and coupons are not used.

```python
plt.figure(figsize=(16,8))
plt.title('Comparison between Departure and arrival Avg population ')
sb.lineplot(data=af['Departure_avg_population'],label='departure')
sb.lineplot(data=af['Arrival_avg_population'],label='arrival')
```

```python
plt.figure(figsize=(16,8))
plt.title('Comparison between Departure and arrival income ')
sb.lineplot(data=af['Departure_avg_INCOME'],label='departure')
sb.lineplot(data=af['Arrival_avg_INCOME'],label='arrival')
```

### As the population for arrival is less on a comparison with departure the fare price drop or rise are also comparitively lesser.

```python
plt.figure(figsize=(20,10))
plt.title('Comparison between Coupon usage by Arrival and Departure population')
sb.lineplot(data=af['Departure_avg_population'],label='departure')
sb.lineplot(data=af['COUPON'],label='coupon')
```

```python
plt.figure(figsize=(20,10))
plt.title('Comparison between Coupon usage by Arrival and Departure population')
sb.lineplot(data=af['Arrival_avg_INCOME'],label='arrival')
sb.lineplot(data=af['COUPON'],label='coupon')
```

### No Coupons have been used by arrival or Departure population.
